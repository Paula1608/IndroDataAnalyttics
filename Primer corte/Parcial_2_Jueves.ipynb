```python


import pandas as pd

teamstats = pd.read_csv('teamstats.csv')
appearances = pd.read_csv('appearances.csv')
games = pd.read_csv('games.csv')
leagues = pd.read_csv('leagues.csv')
players = pd.read_csv('players.csv', encoding='latin-1')
shots = pd.read_csv('shots.csv')


if 'season' in games.columns:
  season_counts = games['season'].value_counts().sort_index()
  print("Número de partidos por temporada:\n", season_counts)

  import matplotlib.pyplot as plt
  season_counts.plot(kind='bar')
  plt.title('Número de partidos por temporada')
  plt.xlabel('Temporada')
  plt.ylabel('Número de partidos')
  plt.show()
else:
  print("La columna 'season' no se encuentra en el DataFrame 'games'.")

if 'league_id' in games.columns and 'name' in leagues.columns:
  league_season = pd.merge(games, leagues, left_on='league_id', right_on='id', how='left')
  league_counts = league_season.groupby('season')['name'].nunique().sort_index()
  print("\nNúmero de ligas diferentes por temporada:\n", league_counts)

  # Visualización (opcional)
  league_counts.plot(kind='bar')
  plt.title('Número de ligas diferentes por temporada')
  plt.xlabel('Temporada')
  plt.ylabel('Número de ligas')
  plt.show()
else:
    print("Las columnas necesarias para el análisis de ligas no están disponibles.")
```

#Calidad de datos



```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



def analyze_data_quality(df, df_name):
    """
    Analiza la calidad de los datos de un DataFrame, buscando valores atípicos y nulos.

    Args:
        df: El DataFrame a analizar.
        df_name: El nombre del DataFrame (para mostrar en los resultados).
    """
    print(f"\nAnálisis de calidad de datos para {df_name}:\n")


    null_counts = df.isnull().sum()
    print("Valores nulos por columna:\n", null_counts)


    for col in df.select_dtypes(include=['number']):
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot de {col} en {df_name}")
        plt.show()


        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        print(f"Valores atípicos en {col}:")
        print(outliers[[col]])
        print("-" * 20)


analyze_data_quality(teamstats, 'teamstats')
analyze_data_quality(appearances, 'appearances')
analyze_data_quality(games, 'games')
analyze_data_quality(leagues, 'leagues')
analyze_data_quality(players, 'players')
analyze_data_quality(shots, 'shots')
```

```python
# prompt: Tendencias: ¿Hay una tendencia en goles por partido u otra métrica relevante con el
#  tiempo?

# Tendencias: ¿Hay una tendencia en goles por partido u otra métrica relevante con el tiempo?

import pandas as pd
import matplotlib.pyplot as plt

# ... (código anterior) ...

# Combinar DataFrames relevantes (games y shots)
games_shots = pd.merge(games, shots, on='game_id', how='left')

# Calcular goles por partido por temporada
goals_per_game_by_season = games_shots.groupby('season')['is_goal'].sum() / games_shots.groupby('season')['game_id'].nunique()

# Visualizar la tendencia de goles por partido
plt.figure(figsize=(10, 6))
goals_per_game_by_season.plot(marker='o')
plt.title('Tendencia de Goles por Partido por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Goles por Partido')
plt.grid(True)
plt.show()


# Análisis adicional:  Otros indicadores relevantes
# - Tiros por partido:
shots_per_game_by_season = games_shots.groupby('season')['shot_id'].count() / games_shots.groupby('season')['game_id'].nunique()
plt.figure(figsize=(10,6))
shots_per_game_by_season.plot(marker='o')
plt.title('Tendencia de Tiros por Partido por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Tiros por Partido')
plt.grid(True)
plt.show()


# - Precisión de tiros (goles / tiros):
shooting_accuracy_by_season = goals_per_game_by_season / shots_per_game_by_season
plt.figure(figsize=(10,6))
shooting_accuracy_by_season.plot(marker='o')
plt.title('Precisión de Tiros por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Precisión de Tiros')
plt.grid(True)
plt.show()

# ... (resto del código) ...
```

```python
# Calcular goles totales por partido
games["total_goals"] = games["homeGoals"] + games["awayGoals"]

# Agrupar por temporada y calcular el promedio de goles
goals_per_season = games.groupby("season")["total_goals"].mean().reset_index()
goals_per_season.rename(columns={"total_goals": "avg_goals_per_game"}, inplace=True)

# Verificamos la tabla
goals_per_season.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.lineplot(data=goals_per_season, x="season", y="avg_goals_per_game", marker="o", linewidth=2)
plt.title("Tendencia de Goles por Partido por Temporada")
plt.xlabel("Temporada")
plt.ylabel("Promedio de Goles por Partido")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
